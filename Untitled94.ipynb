```python
import pandas as pd
import numpy as np

# Function to remove outliers using IQR for all numerical columns
def remove_outliers_iqr(df):
    for column in df.select_dtypes(include=[np.number]).columns:
        # Calculate Q1 (25th percentile) and Q3 (75th percentile)
        Q1 = df[column].quantile(0.25)
        Q3 = df[column].quantile(0.75)
        
        # Calculate the IQR
        IQR = Q3 - Q1
        
        # Define lower and upper bounds for outliers
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        
        # Filter the data to remove outliers
        df = df[(df[column] >= lower_bound) & (df[column] <= upper_bound)]
    
    return df

# Example usage: Remove outliers from the dataset
train_no_outliers = remove_outliers_iqr(train)

# Check the data after removing outliers
train_no_outliers.describe()
```

```python
from sklearn.preprocessing import MinMaxScaler, StandardScaler

# Initialize MinMaxScaler for Normalization
min_max_scaler = MinMaxScaler()

# Initialize StandardScaler for Standardization
standard_scaler = StandardScaler()

# Normalize and Standardize all numerical columns
def normalize_and_standardize(df):
    # Select only numerical columns
    numerical_cols = df.select_dtypes(include=[np.number]).columns
    
    # Apply MinMax Scaling (Normalization)
    df[numerical_cols] = min_max_scaler.fit_transform(df[numerical_cols])
    
    # Apply Standard Scaling (Standardization)
    df[numerical_cols] = standard_scaler.fit_transform(df[numerical_cols])
    
    return df

# Apply the normalization and standardization
train_scaled = normalize_and_standardize(train)

# Check the first few rows of the scaled data
train_scaled.head()
```

```python
import seaborn as sns

# Function to plot confusion matrix
def plot_confusion_matrix(cm, model_name):
    plt.figure(figsize=(6, 4))
    sns.heatmap(cm, annot=True, fmt="d", cmap="Blues", xticklabels=["Not Survived", "Survived"], yticklabels=["Not Survived", "Survived"])
    plt.title(f"Confusion Matrix - {model_name}")
    plt.xlabel("Predicted")
    plt.ylabel("Actual")
    plt.show()

# Plot confusion matrices for each model
plot_confusion_matrix(logreg_cm, "Logistic Regression")
plot_confusion_matrix(svm_cm, "SVM")
plot_confusion_matrix(xgb_cm, "XGBoost")
```
